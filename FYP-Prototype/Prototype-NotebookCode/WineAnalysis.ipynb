These next lines are just a quick clean and sorting of the dataset to make it useable for the prototype.

```python
data = pd.read_csv('./data/winemag-data-130k-v2.csv')
```

```python
data.head()
```

```python
years = data.title.str.extract('([1-2][0-9]{3})').astype('float64')
data = data.assign(year = years)
```

```python
data['Age'] = 2019 - data['year']
```

```python
data.head()
```

```python
data = data.drop(columns=['country','designation','province','region_1','region_2','taster_name','taster_twitter_handle'])
```

```python
data = data.drop(columns=['Unnamed: 0', 'description', 'variety', 'title', 'variety', 'winery', 'year'])
```

```python
data.fillna(data.mean(), inplace=True)
```

```python
data = data.round(0)
data
```

```python
data = data.astype(int)
```

```python
data = data[['points','Age','price']]
data.head()
```

These next few lines are the KNN algorithm being put to work. We seperate our dataset into the needed columns.

We put points and Age into the X axis and then price into the y axis.

```python
X = np.array(data.drop(['price'], 1))
y = np.array(data['price'])
```

Here we seperate the data into 80% training and 20% test.
We then Instantiate the model with 2 neighbors, we have chosen two as this got the most accurate result through testing.
We then fit the model onto our training data and after that we test the models accuracy.

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X,y,test_size=0.2)
 
knn = KNeighborsClassifier(n_neighbors=2)

knn.fit(X_train, y_train)

knn.score(X_test, y_test)
```

We then predict the first 5 rows of prices.

```python
knn.predict(X_test) [0:5]
```

The model is then given input variables. It is given a rating/points of 80 and an Age of 2. It then gives us a predicted price given these variables which is 9.

```python
input_variables = np.array([80,2])
input_variables = input_variables.reshape(1,-1)
```

```python
knn.predict(input_variables)
```

Naming the model 'predictions'.

```python
predictions = knn.predict(X_test)
```

```python
predictions
```

This is where we export the predictions and a file that compares the predictions to the actual prices to a .CSV

```python
prediction = pd.DataFrame(predictions, columns=['Predictions']).to_csv('prediction.csv')
```
